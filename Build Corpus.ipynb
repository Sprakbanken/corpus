## Hent metadata

Hent ut forslag til liste over bøker. Sett verdien av 'size' til en passende verdi.



```python
korpus = nb.book_urn(author='hamsun%knut%', limit=500)
```

## Konverterer til dataramme for å lagre til csv og/eller excel

Samtidig utvides metadataene så det blir litt mer å sortere på.

```python
metakorpus = nb.frame(nb.metadata(korpus), "urn forf år tittel undertittel forlag sjanger sesamid språk oversatt".split())
```

## Bearbeid korpuset

Notebook-systemet er ikke så godt egnet til å bearbeide korpus, så vi lagrer det som en excelfil, og bruker excel til å redigere. Kommandoen `save_frame_to_excel` benytter de innebygde mekanismene i `pandas` for å lagre filen, men har lagt til en sjekk på om filen finnes. Etter lagring kan filen tas til Excel og berbeides der.

### Lagre metadata

```python
nb.save_frame_to_excel(metakorpus, 'metakorpus1.xls')
```

### Hent metadata etter bearbeiding i excel

```python
nyttkorpus = nb.restore_metadata_from_excel('metakorpus1.xls')
```
